### 4.1. Bảng tham khảo điều chỉnh tên nghề

- <= 2 năm ==> Junior
- <= 3 năm ==> Middle
- vị trí > 3 ==> Senior

Để chi tiết hơn, ta phần các nghề thành các cấp như nghề bình thường: <b> Intern < Fresher < Junior < Middle < Senior </b>

### 8.2. Split để gãy gọn và ra cột có ý nghĩa

Giờ chúng ta thấy dữ liệu nhìn còn rất ngu, và với dữ liệu lộn xộn thế này thì chúng ta chỉ muốn lấy giờ thôi. Ta sẽ tư duy theo kiểu: <b>Muốn lấy giá trị nào</b>.
Có khá nhiều giá trị, song có 2 điểm đáng quan trọng chúng ta cần xem xét nhất đối với một thời gian làm việc công ty (cái này tùy theo mỗi cá nhân), đó là:
   - <b>Có phải đi làm cuối tuần không? (thứ 7)</b>: Để giải quyết bài toán 'cuối tuần đi làm', ta sẽ tìm xem trong ô giá trị có cụm 'nghỉ thứ 7', hoặc 'Nghỉ Thứ Bảy', 'cách tuần'... các cách viết khác nhau để xem công ty có yêu cầu làm một thứ 7 nào đó không. Sau đó, ta sẽ trả về một <b>cột 'dummy' 1 tí (dạng 0, 1), nếu công ty làm thứ 7 thì là 0, không làm thứ 7 thì là 1.</b>
   - <b>Một ngày làm việc bao tiếng?</b>: phần này chúng ta sẽ tách các đoạn 'từ:... đến...' để biết một công ty một ngày làm việc bao lâu, từ đó ta sẽ xem một tuần ở công ty đó phải làm việc bao nhiêu tiếng?

Ngoài ra, nếu split theo '0)' cần để ý tới một yếu tố, rằng một vài công ty cho cả thời gian làm thứ 7 vào cạnh thứ 2-6, ví dụ: 'Thứ 2 - Thứ 6 (từ 08:30 đến 18:00), Thứ 7 (từ 08:30 đến 12:00)'. Thế này tức là nó sẽ <b>sinh ra 2 giá trị sau khi split, mà mình chỉ cần 1 thôi. 

Vì vậy, nên để thuận tiện thì:
- làm thứ 7 chúng ta tự động tính số giờ làm là: <b>(ca sáng + cả ngày)/2, hay trung bình của 4 (mặc định ca sáng làm 4h) + 8 (số giờ làm cả ngày), tương đương với 6h.
- công ty không ghi chú thời gian làm tự động tính là 8h/ngày

#### 8.2.1. Tách nội dung đi làm thứ 7 thành một cột riêng

Nhìn tổng quan, ta thấy một công việc đi làm thứ 7 sẽ có các dấu hiệu như sau: <b>Thứ 7 (, Thứ7:, cách tuần, thứ 7 (, T7 đầu tiên, 02 thứ 7, sáng T7, sáng thứ 7

Ta cho hết đống kia vào một list rồi xét xem ô có giá trị này hay không thôi, như đã nói thì <b> có là 0, không là 1

```python
#lập danh sách có dấu hiệu đi làm thứ 7
dumb_working_day = ['Thứ 7 (', 'Thứ7:', 'cách tuần', 'thứ 7 (', 'T7 đầu tiên', '02 thứ 7', 'sáng T7', 'sáng thứ 7']
```

```python
work_saturday = [] # tạo một list để chứa giá trị
for j in df_domain_normalized['job_working_hour']:
    if any(i.lower() in j.lower() for i in dumb_working_day): # phần này lower để hạn chế lỗi nếu nó không tìm được ký tự có case viết hoa
        work_saturday.append(0)  # chứa từ khóa -> làm thứ 7
    else:
        work_saturday.append(1)  # không có từ khóa -> nghỉ thứ 7
```

Task này thì vất cho AI làm cho nhanh, cứ prompt là <b>'tính khoảng cách từ bến xe Giáp Bát tới các quận thuộc Hà Nội, 
chuyển sang dạng dict trong python'</b> để tí nữa tìm cho dễ.

```python
#Khoảng cách từ bến xe Giáp Bát tới các quận/huyện ngoại thành Hà Nội
distances_to_district = {
    "Ba Đình": 8.0,
    "Hoàn Kiếm": 8.3,
    "Đống Đa": 6.0,
    "Hai Bà Trưng": 4.5,
    "Thanh Xuân": 4.0,
    "Hoàng Mai": 3.0,
    "Cầu Giấy": 7.0,
    "Tây Hồ": 10.0,
    "Long Biên": 9.0,
    "Hà Đông": 12.0,
    "Bắc Từ Liêm": 13.0,
    "Nam Từ Liêm": 11.0,
    "Đông Anh": 19.0,
    "Gia Lâm": 14.0,
    "Thanh Trì": 5.0,
    "Sóc Sơn": 24.0,
    "Đan Phượng": 16.0,
    "Hoài Đức": 18.0,
    "Mê Linh": 22.0,
    "Phúc Thọ": 28.0,
    "Ba Vì": 32.0,
    "Phú Xuyên": 30.0,
    "Thạch Thất": 35.0,
    "Quốc Oai": 30.0,
    "Thanh Oai": 26.0,
    "Thường Tín": 25.0,
    "Ứng Hòa": 36.0,
    "Mỹ Đức": 38.0,
    "Chương Mỹ": 40.0,
    "Sơn Tây": 45.0
}
```

```python
district = list(distances_to_district.keys()) #lấy ra list các quận có trong danh sách, lát map lại theo đây để ra khoảng cách 
district
```

```python
#tạo hàm để lát apply
def distance_km(address): 
    for district_name in district: 
        if district_name.lower() in address.lower(): 
            return distances_to_district[district_name]
    return np.nan
```

```python
a = df_working_hour_normalized['job_address'].apply(distance_km)
df_working_hour_normalized['distance_to_work'] = a
df_working_hour_normalized['distance_to_work'] #tạo cột mới
```

### 11.1. 'has_DA_skills' & 'has_scrum'

```python
#2 cái loại skill cần từ điển tí tìm cho dễ
da_skills = ['phân tích dữ liệu', 'trực quan hóa dữ liệu', 'excel', 'sql', 'truy xuất dữ liệu', 'mô hình hóa dữ liệu', 'báo cáo', 'data analysis']
scrum_skills = ['scrum_master', 'scrum']
```

```python
def job_with_da_skill(job_skill):
    for da_skill in da_skills: 
        if da_skill in job_skill: 
            return 1
    return 0
```

```python
a = df_hiring_normalized['job_skills'].apply(job_with_da_skill)
df_hiring_normalized['has_DA_skills'] = a 
df_hiring_normalized['has_DA_skills'] #kỹ năng có DA skill
```

```python
def job_with_scrum(job_skill):
    for scrummy in scrum_skills: 
        if scrummy in job_skill: 
            return 1
    return 0
```

```python
b = df_hiring_normalized['job_skills'].apply(job_with_scrum)
df_hiring_normalized['has_scrum'] = b 
df_hiring_normalized['has_scrum']
```
